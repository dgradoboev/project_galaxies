```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Считываем все данные
galist_2 = pd.read_csv('galist_2.csv')
galist_s = pd.read_csv('galist_s.csv')
grlist_2 = pd.read_csv('grlist_2.csv')
grlist_s = pd.read_csv('grlist_s.csv')
```

```python
#Посмотрим на них
display(galist_2.head(10))
display(galist_s.head(10))
display(grlist_2.head(10))
display(grlist_s.head(10))
```

```python
#Объединяем по стобцу iGrID
g_2 = pd.merge(galist_2, grlist_2, on='iGrID')
g_s = pd.merge(galist_s, grlist_s, on='iGrID')
display(g_2.head(10))
display(g_s.head(10))
```

```python
#Оставляем группы с Ntot>30
g_2 = g_2[g_2['Ntot'] > 30]
g_s = g_s[g_s['Ntot'] > 30]
display(g_2.head(10))
display(g_s.head(10))
```

На сайте с данными выясняем, что:

*REJ2000_# и DEJ2000_# - координаты.*

*z_#, z_# - красное смещение*

```python
def drawPlot(data, GrID, ax, c):
    gr = data[data.iGrID == GrID] #Оставляем только нужную нам группу
    x = np.sqrt((gr.RAJ2000_x - gr.RAJ2000_y)**2 + (gr.DEJ2000_x - gr.DEJ2000_y)**2) #Считаем вектор расстояний
    y = gr.z_x - gr.z_y #Считаем вектор разниц красных смещенний
    ax.scatter(x, y, c=c, marker='x')
    ax.set_xlabel('distance from center in degrees')
    ax.set_ylabel('difference of redshift')
    ax.set_title('iGrID = %d' % (GrID))
    ax.set_ylim(top=max(y)*1.1, bottom=min(y)*1.1)
```

```python
fig, axs = plt.subplots(7, 3, figsize=(15,25))
GrIDs = g_2.iGrID.unique() #Создаём вектор всех групп
for i in range(21):
    drawPlot(g_2, GrIDs[i], axs[i//3, i%3], 'r') #Для каждой группы вызываем функцию и передаём ей нужный 'подграфик'
plt.tight_layout()
```

```python
#Повторим для 2 таблицы
fig, axs = plt.subplots(7, 3, figsize=(15,25))
GrIDs = g_s.iGrID.unique()
for i in range(21):
    drawPlot(g_s, GrIDs[i], axs[i//3, i%3], 'g')
plt.tight_layout()
```

```python
#Сохраним объединенные данные
g_2.to_csv('g_2.csv', index=False)
g_s.to_csv('g_s.csv', index=False)
```

**В TOPCAT загружаем эти файлы обязательно с маркером Table Format - csv. Чтобы разбить данные на группы, идём в Display row subsets, затем Add new mutually exclusive subsets based on automatic classification of column contents, выбираем iGrID за Classification value и выбираем количество категорий (сколько значений iGrID мы отделяем). Теперь при построении графиков мы можем перейти в подмножества и выбрать нужные нам, теперь можно отличать группы и убирать/добавлять их при работе с графиками.**
